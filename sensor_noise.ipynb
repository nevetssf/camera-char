# Definition of ISO

ISO for a sensor is defined in two ways: either by an SNR target, or by the light power required to reach saturation of the detector with the applied gain. 

https://www.dxomark.com/glossary/iso-speed/
https://en.wikipedia.org/wiki/Lumen_(unit)

$H_{sat}$ is the exposure (in $lux \cdot s$) required to achieve saturation of the detector. Actually, it's usually the exposure to achieve 80% of the well capacity, to allow for some highlight recovery headroom. 

The ISO is defined by $S_{sat} = \frac{78}{H_{sat}}$



For example if $S_{sat}=100$, then $H_{sat} = 0.78 lux \cdot s$. 

$1 lux =1 lm/m^2$

Irradience is in $W/m^2$. Conversion from luminance to irradiance depends on the wavelength. 

For a given pixel size $p$ the total integrated power to reach saturation is 

This reference https://www.hhyu.org/posts/photons/ says 1 lumen is 4.09*10 photons/s at the peak flux of 555 nm.

How many photos saturate the detector, and what is the final SNR?

$$
\begin{aligned}
H_{sat} &= \frac{78}{S_{sat}} \mathrm{lux} \cdot \mathrm{s} \\
&= \frac{78}{S_{sat}} \mathrm{lumen}/\mathrm{m}^2/\mathrm{s} \\
&= \frac{78}{S_{sat}} 4.09\times10^{15} \mathrm{photon}/\mathrm{m}^2 \\
&= \frac{3.19 \times10^{17}}{S_{sat}} \mathrm{photon}/\mathrm{m}^2 \\
\end{aligned}
$$
